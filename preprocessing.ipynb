We can see some unknown values like 'Insgesamt' and '95 u. mehr' in 'age' column

```python
# 'Insgesamt' in english means all in all or total

data[data.age == 'Insgesamt'].sample(5)
```

We can see the ages with 'Insgesamt' have very large values.

```python
#so we can drop them since we are working with age ranges

data = data[data.age != 'Insgesamt']
```

```python
data.head()
```

```python
# '95 u. mehr' in english means '95 and more'
# we can replace 'u. mehr' with '+'

data['age'] = data['age'].str.replace(' u. mehr', '+')
```
